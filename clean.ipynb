# Getting started with cleaning 

## Rename, drop, merge columns

```python
# since columns are non existing, add column names
df.columns = ['ID', 'as in', 'Name', 'Helpful', 'Helpful out of...', 'Review Text I', 'Review Text II', 'Review Text III', 'Rating', 'Summary', 'Time', 'Day', 'Year']
df.head()
```

```python
# drop as in column, helpful column, helpful out of... column by setting dataframe equal to dataframe and its columns
df = df [['ID', 'Name',  'Review Text I', 'Review Text II', 'Review Text III', 'Rating', 'Summary', 'Time', 'Day', 'Year']]
df.head()
```

```python
# Merge review columns into one for better readability
df['Review'] = df['Review Text I'] + df['Review Text II'] + df['Review Text III']
df = df[['ID', 'Name',  'Review', 'Rating', 'Summary', 'Time', 'Day', 'Year']]
df.head()
```

## Split unnecessary characters / repetitive info off each column using Regex and .split()

### Proceed column by column as they are all different - cannot use just regex, may need split prior

```python
splitid = df.ID.str.split(':')
df.ID = splitid.str.get(1) 
df.ID = df.ID.str.replace('["]', '', regex=True)
df.head()
```

```python
splitname = df.Name.str.split(':')
df.Name = splitname.str.get(1) 
df.Name = df.Name.str.replace('["]', '', regex=True)
df.head()
```

```python
splitreview = df.Review.str.split(':')
df.Review = splitreview.str.get(1) 
df.Review = df.Review.str.replace('["]', '', regex=True)
df.head()
```

```python
splitrating = df.Rating.str.split(':')
df.Rating = splitrating.str.get(1) 
df.Rating = df.Rating.str.replace('["]', '', regex=True)
df.head()
```

```python
splitsummary = df.Summary.str.split(':')
df.Summary = splitsummary.str.get(1) 
df.Summary = df.Summary.str.replace('["]', '', regex=True)
df.head()
```

```python
df.Time = df.Time.str.replace('[reviewTimeunxR["}:]', '', regex=True)
df.head()
```

```python
df.Day = df.Day.str.replace('["reviewTime":}]', '', regex=True)
df.head()
```

```python
df.Year = df.Year.str.replace('["}]', '', regex=True)
df.head()
```
